### Subsetting human cells and pre-processing

From our species calling notebook, we defined a list of clusters that are human/malignant. We will label the species, subset them, and analyze them separately for our subsequent analysis. 

```python
# List of clusters corresponding to human_cancer cells
human_cancer_clusters = [0, 2, 4, 5, 6, 7, 8, 14, 15, 19, 46, 47, 51, 52]

# Create a new column `human_cancer` in adata.obs
adata.obs['human_cancer'] = adata.obs['leiden'].astype(int).isin(human_cancer_clusters)
```

```python
adata.obs
```

```python
# Create the `mouse_normal` column, which is the opposite of `human_cancer`
adata.obs['mouse_normal'] = ~adata.obs['human_cancer']

adata.obs
```

```python
# only cluster 51 and 52 are likely imaging artefacts (e.g., speckle), as you can see their spatial distribution is restricted to a particular fov(s). 
# The safest way to address this is removing them from our dataset. 
mask = ~adata.obs["leiden"].isin(["51", "52"])

# Subset and copy
adata_clean = adata[mask].copy()
```

```python
adata_clean
```

```python
# want to add in mouse and human labels for genes in the adata.var
adata_clean.var
```

```python
import numpy as np

adata_clean.var["gene_species"] = np.where(
    adata_clean.var_names.str.endswith("--H1"),
    "human",
    "mouse"
)
```

```python
adata_clean.obs["species"] = np.where(
    adata_clean.obs["human_cancer"] == True,
    "human",
    "mouse"
)
```

### Pre-processing on human cells 
Now that we have human cells, we will do some basic pre-processing: filtering on volume, counts normalization, leiden clustering, UMAP, etc. 

```python
adatah.obs["volm"]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.hist(adatah.obs["volm"].dropna(), bins=50)
plt.axvline(25000, linestyle="--", linewidth=2)
plt.xlabel("volm")
plt.ylabel("Count")
plt.title("Histogram of adatah.obs['volm']")
plt.show()
```

```python
### there is a large amount of cells with small volumes, which may be technical artefacts and can impact downstream analyis. As such, we will filter all cells less than 25,000
adatah = adatah[adatah.obs["volm"] > 25000].copy()
adatah
```

```python
# ensure raw counts before normalization
adatah.X
```

```python
sc.pp.calculate_qc_metrics(adatah, percent_top=None, inplace=True)
sc.pp.normalize_total(adatah, target_sum=np.median(adatah.obs["total_counts"]))
sc.pp.log1p(adatah)
```

```python
# parameters

sc.pp.pca(adatah)
sc.pp.neighbors(adatah)  
sc.tl.umap(adatah,random_state=9,min_dist=0.2)
sc.tl.leiden(adatah, resolution=.4)
```
